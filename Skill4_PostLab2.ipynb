Program 2: Duplicate Handling & Data Summary
Write a Python program to: Identify duplicate rows using duplicated(). Remove
duplicates.Display summary information using info() and describe().

```python
#Step 1: Create a Sample DataFrame with Duplicates
import pandas as pd

data = {
    "Name": ["Amit", "Neha", "Ravi", "Amit", "Neha", "Kiran"],
    "Age": [20, 21, 19, 20, 21, 22],
    "Marks": [85, 90, 78, 85, 90, 88]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
df
```

```python
# Step 2: Identify Duplicate Rows
print("Duplicate Rows (True means duplicate):")
df.duplicated()
```

```python
# Step 3: Remove Duplicate Rows
df_unique = df.drop_duplicates()

print("DataFrame after removing duplicates:")
df_unique
```

```python
# Step 4: Display Summary Information
print("Information about DataFrame:")
df_unique.info()

print("\nStatistical Summary:")
df_unique.describe()
```
